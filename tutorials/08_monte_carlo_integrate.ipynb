## Integration using Monte-Carlo method

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Function to integrate

```python
def func(x):
    return np.log(x)
```

### Plotting the function

```python
x_min, x_max = 1, 4
x_array = np.linspace(x_min, x_max, 50)
plt.plot(x_array, func(x_array))
plt.xlabel('x', fontsize=14)
plt.ylabel('log(x)', fontsize=14)
plt.show()
```

### Monte Carlo integration
$$ Area = \int_{x_{min}}^{x_{max}} log(x).dx = \frac{N_{inside}}{N_{total}}A_{box} $$

```python
y_min, y_max = 0, 2.0
count_total = 4000
count_inside = 0
# saving points for plotting in a python list
random_number_inside_x = [] # initializing a python list
random_number_inside_y = []
random_number_outside_x = []
random_number_outside_y = []
for i in range(count_total):
    random_number_x = np.random.uniform(x_min, x_max) # returns a random float between x_min, x_max
    random_number_y = np.random.uniform(y_min, y_max) # returns a random float between y_min, y_max
    # checking if point lies inside the integral domain
    if random_number_y - func(random_number_x) < 0:
        count_inside += 1
        random_number_inside_x.append(random_number_x)
        random_number_inside_y.append(random_number_y)
    else:
        random_number_outside_x.append(random_number_x)
        random_number_outside_y.append(random_number_y)

integral = (count_inside/count_total)*(x_max - x_min)*(y_max - y_min)
print("Integral = ", integral) # 2.5452
plt.scatter(random_number_inside_x, random_number_inside_y, facecolors='none', edgecolors='r')
plt.scatter(random_number_outside_x, random_number_outside_y, facecolors='none',  edgecolors='g')
plt.show()
```
